##3. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
```
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017
The code snippet gives the output shown in the following screenshot:
```



image.png

````python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

# Creating a sample DataFrame
data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Open': [774.25, 776.030029, 779.309998, 779, 779.659973],
    'High': [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}

df = pd.DataFrame(data)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Set 'Date' column as the index
df.set_index('Date', inplace=True)

# Plotting
plt.figure(figsize=(10, 6))

# Plotting the 'Open', 'High', 'Low', and 'Close' prices
plt.plot(df.index, df['Open'], marker='o', linestyle='-', color='r', label='Open')
plt.plot(df.index, df['High'], marker='o', linestyle='-', color='g', label='High')
plt.plot(df.index, df['Low'], marker='o', linestyle='-', color='b', label='Low')
plt.plot(df.index, df['Close'], marker='o', linestyle='-', color='purple', label='Close')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Alphabet Inc. Financial Data (Sample)')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
````

##5 Write a Python program to plot quantities which have an x and y position. The code snippet gives the output shown in the following screenshot:

````python
import matplotlib.pyplot as plt

# Sample data
x1_positions = [1, 2, 3, 4, 5]
y1_positions = [10, 5, 20, 15, 25]
x_positions = [5, 7, 8, 9, 10]
y_positions = [15, 10, 25, 15, 30]

# Plotting quantities with x and y positions and custom markers
plt.figure(figsize=(8, 6))

plt.scatter(x_positions, y_positions, color='green', marker='*', label='y_positions')  # '*' for y_positions
plt.scatter(x1_positions, y1_positions, color='blue', marker='o', label='x_positions')   # 'o' for x_positions

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quantities with X and Y Positions')

# Adding legend
plt.legend()

# Display the plot
plt.show()

````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt

# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(languages, popularity, color='b')

# Adding labels and title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.grid()

# Display the plot
plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
df.info
````

````python
import pandas as pd


# Define the list of products
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Assuming month_number is a continuous sequence starting from 1
# Find the last month in the DataFrame
last_month = df['month_number'].max()

# Filter the DataFrame for the last year
last_year_data = df[df['month_number'] > last_month - 12]

# Calculate total sales for each product
total_sales_last_year = last_year_data[products].sum()

# Create a pie chart for total sales distribution
plt.figure(figsize=(8, 8))
plt.pie(total_sales_last_year, labels=products, autopct='%1.1f%%', startangle=140, colors=['green', 'blue', 'red', 'yellow', 'blue', 'magenta'])
plt.title('Total Sales Distribution for Last Year')

plt.show()

````

##Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
The Subplot should look like this.
image.png


````python
import pandas as pd
import matplotlib.pyplot as plt


# Extract data for 'bathingsoap' and 'facewash'
bathingsoap_data = df['bathingsoap']
facewash_data = df['facewash']
months = df['month_number']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Plot 'bathingsoap' data
axes[0].plot(months, bathingsoap_data, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
axes[0].set_title('Bathing Soap Sales Over Months')
axes[0].set_xlabel('Month Number')
axes[0].set_ylabel('Units Sold')

# Plot 'facewash' data
axes[1].plot(months, facewash_data, marker='o', color='green', linestyle='-', linewidth=2, markersize=8)
axes[1].set_title('Facewash Sales Over Months')
axes[1].set_xlabel('Month Number')
axes[1].set_ylabel('Units Sold')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import pandas as pd
import matplotlib.pyplot as plt

# Extract data for all products
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
sales_data = df[products]

# Create a stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(df['month_number'], sales_data.T, labels=products, colors=['green', 'blue', 'red', 'yellow', 'blue', 'magenta'])
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')

plt.show()
````
